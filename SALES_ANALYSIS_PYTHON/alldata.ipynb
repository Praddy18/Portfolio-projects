```python
#####Which city had highest number of sales?######
```

```python
###Lets add city columns
```

```python
df['ct'] = df['Purchase Address'].apply(lambda x : x.split(',')[1])
df['state'] = df['Purchase Address'].apply(lambda x : x.split(',')[2].split(' ')[1])
df['City'] = df['ct'].map(str) + ' ' + df['state'].map(str)

```

```python
df.head()
```

```python
df.drop(['ct', 'state'], axis=1)
```

```python
df.head()
```

```python
cr=df.groupby('City').sum()
```

```python
cr
```

```python
cities = [city for city,df in df.groupby('City')] ##for y plotting order matters 
plt.bar(cities,cr['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.xlabel('City names')
plt.ylabel('Sales in usd')
plt.show()
```

```python
## What time should we consider to start offers to boost our sales
```

```python
df['Order Date']= pd.to_datetime(df['Order Date'])
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
```

```python
df
```

```python
hr=df.groupby('Hour').count()
hr
```

```python
hours = [hour for hour,df in df.groupby('Hour')]
plt.plot(hours,hr)
plt.xticks(hours,rotation='vertical',size=8)
plt.grid()
plt.show()
```
